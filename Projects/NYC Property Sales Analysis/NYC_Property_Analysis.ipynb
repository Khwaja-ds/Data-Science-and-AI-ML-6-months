```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\HP\Downloads\Projects\NYC_Property_Analysis\nyc-rolling-sales.csv")
df.head()
```

```python
df.shape
```

```python
# Now here we have to clean the dataset and going to do that
```

```python
# Convert 'SALE PRICE' to numeric
df['SALE PRICE'] = pd.to_numeric(df['SALE PRICE'].replace('-', '0'), errors='coerce')
```

```python
df.head()
```

```python
# Convert 'SALE DATE' to datetime
df['SALE DATE'] = pd.to_datetime(df['SALE DATE'], errors='coerce')
df
```

```python
# Remove entries with missing or zero sale prices
df_cleaned = df[df['SALE PRICE'] > 10000]
df_cleaned
```

```python
# Convert square footage columns to numeric
df_cleaned['LAND SQUARE FEET'] = pd.to_numeric(df_cleaned['LAND SQUARE FEET'], errors='coerce')
df_cleaned['GROSS SQUARE FEET'] = pd.to_numeric(df_cleaned['GROSS SQUARE FEET'], errors='coerce')
df_cleaned
```

```python
# Drop rows with missing values in important columns
df_cleaned = df_cleaned.dropna(subset=['SALE DATE', 'GROSS SQUARE FEET', 'SALE PRICE'])
df_cleaned
```

```python
df_cleaned.info()
```

# EDA

```python
# 1. Sale Price Distribution
plt.figure(figsize=(10,6))
sns.histplot(df_cleaned['SALE PRICE'], bins=100, kde=True)
plt.xlim(0, 10000000)  # Limit to remove extreme outliers
plt.title("Distribution of Sale Prices in NYC")
plt.xlabel("Sale Price")
plt.ylabel("Frequency")
plt.show()
```

```python
#  2. Sales Trend Over Time
monthly_sales = df_cleaned.resample('M', on='SALE DATE').agg({
    'SALE PRICE': 'median',
    'BOROUGH': 'count'
}).rename(columns={'BOROUGH': 'Sales Volume'})

monthly_sales.plot(y=['SALE PRICE', 'Sales Volume'], secondary_y='SALE PRICE', figsize=(14,6), title="Monthly Sales Volume & Median Price")
```

```python
# 3. Top Boroughs by Total Sales
borough_map = {
    1: "Manhattan",
    2: "Bronx",
    3: "Brooklyn",
    4: "Queens",
    5: "Staten Island"
}

df_cleaned['BOROUGH'] = df_cleaned['BOROUGH'].map(borough_map)

plt.figure(figsize=(8,6))
sns.countplot(data=df_cleaned, x='BOROUGH', order=df_cleaned['BOROUGH'].value_counts().index)
plt.title("Number of Sales by Borough")
plt.ylabel("Sales Count")
plt.xlabel("Borough")
plt.show()
```

```python
# 4. Sale Price vs Gross Square Feet 
plt.figure(figsize=(10,6))
sns.scatterplot(data=df_cleaned, x='GROSS SQUARE FEET', y='SALE PRICE', alpha=0.3)
plt.xlim(0, 20000)
plt.ylim(0, 10000000)
plt.title("Sale Price vs Gross Square Feet")
plt.xlabel("Gross Square Feet")
plt.ylabel("Sale Price")
plt.show()
```

```python
# 5. Top Neighborhoods by Average Price 
top_neighborhoods = df_cleaned.groupby('NEIGHBORHOOD')['SALE PRICE'].mean().sort_values(ascending=False).head(10)

top_neighborhoods.plot(kind='bar', figsize=(10,6), title="Top 10 Neighborhoods by Avg Sale Price")
plt.ylabel("Average Sale Price")
plt.xlabel("Neighborhood")
plt.show()
```
