```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots prettier
sns.set(style='whitegrid')
%matplotlib inline
```

```python
matches = pd.read_csv(r"C:\Users\HP\Downloads\IPL_DATA_ANALYSIS\Data\matches.csv")
deliveries = pd.read_csv(r"C:\Users\HP\Downloads\IPL_DATA_ANALYSIS\Data\deliveries.csv")
```

```python
matches.drop(['umpire1', 'umpire2', 'umpire3'], axis=1, inplace=True)
```

```python
season_matches = matches['Season'].value_counts().sort_index()
season_matches.plot(kind='bar', color='skyblue')
plt.title("Matches Per Season")
plt.xlabel("Season")
plt.ylabel("Matches")
plt.show()
```

```python
most_wins = matches['winner'].value_counts().head(10)
sns.barplot(x=most_wins.values, y=most_wins.index, palette='viridis')
plt.title("Top 10 Winning Teams")
plt.xlabel("Wins")
plt.ylabel("Teams")
plt.show()
```

```python
toss_decision = matches['toss_decision'].value_counts()
toss_decision.plot(kind='pie', autopct='%1.1f%%', startangle=90, shadow=True)
plt.title("Toss Decisions")
plt.ylabel("")
plt.show()
```

```python
venue_count = matches['venue'].value_counts().head(10)
sns.barplot(y=venue_count.index, x=venue_count.values, palette='cubehelix')
plt.title("Top 10 Venues")
plt.xlabel("Matches Played")
plt.show()
```

```python
top_batsmen = deliveries.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=top_batsmen.values, y=top_batsmen.index, palette='magma')
plt.title("Top 10 Run Scorers")
plt.xlabel("Runs")
plt.ylabel("Players")
plt.show()
```

```python
wickets = deliveries[deliveries['dismissal_kind'].notnull()]
top_bowlers = wickets['bowler'].value_counts().head(10)
sns.barplot(x=top_bowlers.values, y=top_bowlers.index, palette='Set2')
plt.title("Top 10 Wicket Takers")
plt.xlabel("Wickets")
plt.show()
```
