```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # optional for heatmap
```

```python
df =pd.read_csv(r"C:\Users\HP\Downloads\heart.csv")
```

```python
df.info()
df.isnull().sum()
```

```python
plt.hist(df['age'], bins=15, color='skyblue', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
df['target'].value_counts().plot.pie(labels=["No Disease", "Disease"], autopct="%1.1f%%", colors=['lightgreen', 'salmon'])
plt.title("Heart Disease Proportion")
plt.ylabel("")
plt.show()
```

```python
sns.boxplot(x='target', y='chol', data=df)
plt.title("Cholesterol Levels by Heart Disease Presence")
plt.xlabel("Heart Disease (0 = No, 1 = Yes)")
plt.ylabel("Cholesterol")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Feature Correlation Heatmap")
plt.show()
```
